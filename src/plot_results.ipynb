```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
res = pd.read_csv("sporadic_result.csv")
```

```python
res
```

```python
f, ax = plt.subplots(1, figsize=(16, 9))

to_plot = ["Vestal", "AMCmax"]
for c in to_plot:
    sns.lineplot(data=res, x="target_U", label = c, y=c, ax=ax)
```

```python
f, ax = plt.subplots(1, figsize=(16, 9))

to_plot = ["OCBP", "PLRS", "LPA"]
for c in to_plot:
    sns.lineplot(data=res, x="target_U", label = c, y=c, ax=ax)
```
