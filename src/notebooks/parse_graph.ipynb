```python
p = "../cpp/test_graph.txt"
with open(p) as f:
    graph_str = f.read()
```

```python
graph_t = graph_str.splitlines()
```

```python
node_labels = {}
edges = {}

for e in graph_t[2:-1]:
    if len(e)>0:
        if e.split(" ")[1][:3] == "[la":
            nid = e.split(" ")[0]
            depth, hash = nid.split("_")
            depth = int(depth)
            hash = int(hash)
            if depth in node_labels.keys():
                node_labels[depth][hash] = " ".join(e.split(" ")[1:])
            else:
                node_labels[depth] = {hash: " ".join(e.split(" ")[1:])}
        elif e.split(" ")[1] == "->":

            parent, child = e.split(" -> ")
            _, parent = parent.split("_")
            depth, child = child.split("_")
            depth = int(depth)
            parent = int(parent)
            child = int(child)

            e = f"{parent} -> {child}"

            if depth in edges.keys():
                edges[depth].append(e)
            else:
                edges[depth] = [e]
```

```python
fail_states = set()

for k, v in node_labels[max(node_labels.keys())].items():
    if "black" in v:
        fail_states.add(k)
```

```python
depth = max(node_labels.keys())
```

```python
leafs = list(fail_states)
depth = max(node_labels.keys())
all_states = {depth: set(leafs)}
all_edges = set()

def backtrack_leafs(leafs, all_states, all_edges, depth):

    print(depth)
    
    if depth == 0:
        return all_states, all_edges
    else:
        new_leafs = set()
        for k in leafs:
            for e in edges[depth]:
                parent, child = e.split(" -> ")

                child = int(child)
                parent = int(parent)
                if k == child:
                    new_leafs.add(parent)
                    all_edges.add(e)

        all_states[depth-1]=new_leafs
        
        return backtrack_leafs(new_leafs, all_states, all_edges, depth-1)


all_states, all_edges = backtrack_leafs(leafs, all_states, all_edges, depth)
```

```python
labels = ""
for k in all_states:
    labels += k + " " + node_labels[k]+"\n"


edges_select = '\n'.join(all_edges)

new_dot_file = f"""
digraph G {{
node[shape="box",style="rounded,filled"]
{labels}
{edges_select}
}}
"""
```

```python
import re
pattern = r'\d\d?_(\d+)'
replace = r'\1'
new_dot_file = re.sub(pattern, replace, new_dot_file)
```

```python
with open("g_mwuf_v4.dot", 'w') as f:
    f.write(new_dot_file)
```
